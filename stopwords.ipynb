```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words("english")
```

```python
print(en_stopwords)
```

```python
sentence = 'it was too far to go to the shop and he did not want her to walk'
```

```python
sentence_no_stopwords = ' '.join([words for words in sentence.split() if words not in en_stopwords])
```

```python
print(sentence_no_stopwords)
```

```python
en_stopwords.remove('did')
en_stopwords.remove('not')
```

```python
en_stopwords.append('go')
```

```python
sentence_no_stopwords_custom = ' '.join([words for words in sentence.split() if words not in en_stopwords])
```

```python
print(sentence_no_stopwords_custom)
```
