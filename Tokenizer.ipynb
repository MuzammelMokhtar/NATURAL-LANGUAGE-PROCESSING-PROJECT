```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = 'sam was a great help to me in the store. the cashier was very rude to me, I think her name was eleanor'
```

```python
sent_tokenize(sentences)
```

```python
sentences2 = 'amazing work from sadeen!'
```

```python
word_tokenize(sentences2)
```

```python
sentences3 = 'sarah was able to help me find the items i needed quickly. great service from sarah she found me what i wanted'
```

```python
word_tokenize(sentences3)
```
