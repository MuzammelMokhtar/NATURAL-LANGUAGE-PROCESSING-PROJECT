```python
list = ['Amad start from A','Ahmad start from B']
```

### Optional Pattern ? ###

```python
pattern = r'Ah?mad'
```

```python
for i in list:
    if re.search(pattern,i):
        print(i)
```

### Start Pattern ^ ###

```python
pattern2 = r'^s'
```

```python
list2 = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']
```

```python
for i in list2:
    if re.search(pattern2, i):
        print(i)
```

### End Pattern $ ###

```python
pattern3 = r'!$'
```

```python
for i in list2:
    if re.search(pattern3, i):
        print(i)
```

### Pipe for Or Pattern | ###

```python
pattern4 = r'(need|want)ed'
```

```python
for i in list2:
    if re.search(pattern4, i):
        print(i)
```

### Square Bracket [] A set of characters we want to match

This one for removing punctuation in the sentence
^ for not,
\w for detecting words,
\s for detecting spaces

```python
pattern5 = r'[^\w\s]'
```

```python
list3_without_punctuation = []
```

```python
for i in list2:
    new_string = re.sub(pattern5, '', i)
    list3_without_punctuation.append(new_string)
        
```
