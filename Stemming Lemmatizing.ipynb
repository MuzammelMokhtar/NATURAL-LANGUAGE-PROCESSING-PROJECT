```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
list = ['connected', 'connecting', 'connects', 'connect', 'connectivity']
```

```python
for i in list:
    print(i,": ", ps.stem(i))
```

```python
list2 = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
for i in list2:
    print(i,": ", ps.stem(i))
```

```python
list3 = ['likes', 'better', 'worse']
```

```python
for i in list3:
    print(i,": ", ps.stem(i))
```
