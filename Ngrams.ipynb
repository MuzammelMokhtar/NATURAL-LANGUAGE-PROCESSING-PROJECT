### Ngrams help recognize relationship between words

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'cashier', 'was', 'very', 'rude', 'to', 'me', 'I', 'think', 'her', 'name', 'was', 'eleanor']
```

```python
unigrams = pd.Series(nltk.ngrams(tokens, 1)).value_counts() 
```

```python
unigrams
```

```python
unigrams.sort_values().plot.barh(color = 'lightsalmon', width = .8, figsize = (5,3))
plt.title('My first plot is Unigrams')
```

```python
bigrams = pd.Series(nltk.ngrams(tokens, 2)).value_counts()
```

```python
bigrams
```

```python
trigrams = pd.Series(nltk.ngrams(tokens, 3)).value_counts()
```
